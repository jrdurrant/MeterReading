# Visualize Synthetic Meter Displays

Our synthetic data generation is represented as a "dataset" object that when accessed will generate the image

For simplicity we treat the meter values as being integers (and then convert back to decimals later on as necessary). For example a reading of `46168.5` is accessed from the dataset as `dataset[461685]`.

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torchvision

from data import DisplayDataset
```

```python
dataset = DisplayDataset()
```

We can generate images for any number, so pick some at random and see what the synthetic images look like

```python
meter_values = torch.randint(0, len(dataset), [4])

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
for i, meter_value in enumerate(meter_values.tolist()):
    image_tensor, metadata = dataset[meter_value]
    image = torchvision.transforms.functional.to_pil_image(image_tensor)
    axes.flatten()[i].imshow(np.asarray(image))
    axes.flatten()[i].set_title(f'Meter value = {meter_value / 10}')
```
